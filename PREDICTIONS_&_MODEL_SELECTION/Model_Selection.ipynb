```python
file_name = "29 - Prediction_0.3_Normal"
split = file_name.split("_")
df = pd.read_csv(file_name+".csv")
df.rename({'Unnamed: 0': 'Index'}, axis=1, inplace=True)
indx = [257,516,138,398,751,1490]
df= df[df.Index.isin(indx)]
df.loc[df["Fetal_Health"] != 1, "Fetal_Health"] = "Not " + split[2]
df.loc[df["Fetal_Health"] == 1, "Fetal_Health"] = split[2]
df.loc[df["predClass.step"] != 1,"predClass.step"] = "Not " + split[2]
df.loc[df["predClass.step"] == 1, "predClass.step"] = split[2]
df = df[["Index","Fetal_Health","predClass.step"]]
df.rename(columns = {'predClass.step':split[1] + " " + split[2]}, inplace = True)
df = df.reset_index(drop=True)
```

```python
folder = ["29 - Prediction_0.5_Normal",
         "29 - Prediction_0.7_Normal",
             "29 - Prediction_0.3_Suspect",
           "29 - Prediction_0.5_Suspect",
         "29 - Prediction_0.7_Suspect",
          "29 - Prediction_0.3_Pathological",
           "29 - Prediction_0.5_Pathological",
         "29 - Prediction_0.7_Pathological"
         ]



for file_name in folder:
    split = file_name.split("_")
    df_1 = pd.read_csv(file_name+".csv")
    df_1.rename({'Unnamed: 0': 'Index'}, axis=1, inplace=True)
    indx = [257,516,138,398,751,1490]
    df_1= df_1[df_1.Index.isin(indx)]
    df_1.loc[df_1["Fetal_Health"] != 1, "Fetal_Health"] = "Not " + split[2]
    df_1.loc[df_1["Fetal_Health"] == 1, "Fetal_Health"] = split[2]
    df_1.loc[df_1["predClass.step"] != 1,"predClass.step"] = "Not " + split[2]
    df_1.loc[df_1["predClass.step"] == 1, "predClass.step"] = split[2]
    df_1 = df_1[["Index","Fetal_Health","predClass.step"]]
    df_1.rename(columns = {'predClass.step':split[1] + " " + split[2]}, inplace = True)
    df_1 = df_1.reset_index(drop=True)
    pred = df_1.iloc[:,[2]]
    df = df.join(pred)
```
